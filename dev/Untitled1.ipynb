```python
def get_file_and_index(sim_index, num_files=50, sims_per_file=10):
    file_index = sim_index // sims_per_file + 1
    inner_index = sim_index % sims_per_file
    return file_index, inner_index

# Example: To find which file and what index sim[150] is in
file_index, inner_index = get_file_and_index(150)
print(f"sim[150] is in data_{file_index} at index {inner_index}")
```

```python
file_index, inner_index = get_file_and_index(499)
print(f"sim[150] is in data_{file_index} at index {inner_index}")
```

```python
for i in range(499):
    file_index, inner_index = get_file_and_index(i)
    print(f"sim[{i}] is in data_{file_index} at index {inner_index}")
```
